```python
import nltk
from nltk.corpus import stopwords
from nltk.probability import FreqDist

brownwds = [w.lower() for w in nltk.corpus.brown.words() if w.isalpha()]
brownfd = FreqDist(brownwds)
brownfd_nostop = FreqDist(brownwds)

for w in stopwords.words('english'):
    if w in brownfd_nostop.keys():
        brownfd_nostop[w] = 0

print brownfd.items()[:50]
```

```python
#print these most common words into a file
f = open('brown_words_nostop', 'w')
rank = 0
for w in brownfd_nostop.items():
    f.write(w[0] + ' ' + str(rank) + '\n')
    rank += 1
f.close()
```

```python
#Comparison of overall word frequencies

#Analyze the first N words
n = 200

#first, open the files in which all words are stored, and store their words and word ranks.
fwash = open('./wash_words_nostop')
hwash = {} #maps word to word rank
lwash = [] #maps word rank to word.
linecnt = 0
for line in fwash:
    linecnt += 1
    if linecnt > n:
        break
    larr = line.split(' ')
    hwash[larr[0]] = int(larr[1])
    lwash += [larr[0]]

fadams = open('./adams_wds_nostop')
hadams = {} #maps word to word rank
ladams = [] #maps word rank to word.
linecnt = 0
for line in fadams:
    linecnt += 1
    if linecnt > n:
        break
    larr = line.split(' ')
    hadams[larr[0]] = int(larr[1])
    ladams += [larr[0]]

fbrown = open('./brown_words_nostop')
hbrown = {} #maps word to word rank
lbrown = [] #maps word rank to word.
linecnt = 0
for line in fbrown:
    linecnt += 1
    if linecnt > n:
        break
    larr = line.split(' ')
    hbrown[larr[0]] = int(larr[1])
    lbrown += [larr[0]]

#check that this is working...
print lwash[:20]
print ladams[:20]
print lbrown[:20]


#now, print a few different statistics into an output file
fout = open('wordfreq_analysis', 'w')
index = 0
#compare the top words of each list.
fout.write("Most Frequent Words:\n\n")
fout.write("+--------------------+--------------------+--------------------+\n")
fout.write("  Brown              |  Adams             |  Washington \n")
fout.write("+--------------------+--------------------+--------------------+\n")
for index in range(len(lwash)):
    fout.write(lbrown[index] + '\t\t\t' + ladams[index] + '\t\t\t' + lwash[index] + '\n')


    
#compare brown's words with the words of the presidents.
#here, rank n refers to the nth most commonly used word.
fout.write('\n\n\n\n')
fout.write('Word Rank Comparison of Brown Words: \n\n')
fout.write("+--------------------+--------------------+--------------------+\n")
fout.write("  Brown Word         |  Adams             |  Washington \n")
fout.write("+--------------------+--------------------+--------------------+\n")

for index in range(len(lbrown)):
    wd = lbrown[index]
    #write the word.
    fout.write(str(index) + '. ' + wd + '\t\t\t')
    #write rank in adams words
    if wd in ladams:
        fout.write(str(hadams[wd]) + '\t\t\t')
    else:
        fout.write('>200' + '\t\t\t')
    #write rank in wash words
    if wd in lwash:
        fout.write(str(hwash[wd]) + '\n')
    else:
        fout.write('>200' + '\n') 


        
#compare adams's words with washington, Brown.
#here, rank n refers to the nth most commonly used word.
fout.write('\n\n\n\n')
fout.write('Word Rank Comparison of Adams Words: \n\n')
fout.write("+--------------------+--------------------+--------------------+\n")
fout.write("  Adams Word         |  Washington        |  Brown \n")
fout.write("+--------------------+--------------------+--------------------+\n")

for index in range(len(ladams)):
    wd = ladams[index]
    #write the word.
    fout.write(str(index) + '. ' + wd + '\t\t\t')
    #write rank in adams words
    if wd in lwash:
        fout.write(str(hwash[wd]) + '\t\t\t')
    else:
        fout.write('>200' + '\t\t\t')
    #write rank in wash words
    if wd in lbrown:
        fout.write(str(hbrown[wd]) + '\n')
    else:
        fout.write('>200' + '\n') 



#compare washington's words with washington, Brown.
#here, rank n refers to the nth most commonly used word.
fout.write('\n\n\n\n')
fout.write('Word Rank Comparison of Washington Words: \n\n')
fout.write("+--------------------+--------------------+--------------------+\n")
fout.write("  Washington Word    |  Adams             |  Brown \n")
fout.write("+--------------------+--------------------+--------------------+\n")

for index in range(len(lwash)):
    wd = lwash[index]
    #write the word.
    fout.write(str(index) + '. ' + wd + '\t\t\t')
    #write rank in adams words
    if wd in ladams:
        fout.write(str(hadams[wd]) + '\t\t\t')
    else:
        fout.write('>200' + '\t\t\t')
    #write rank in wash words
    if wd in lbrown:
        fout.write(str(hbrown[wd]) + '\n')
    else:
        fout.write('>200' + '\n') 

fout.close()
```
