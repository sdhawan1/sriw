```python
#first, try to filter out all documents that are not letters written by John Adams
#write the results in a file ("writtenby") [never need to run this cell again]
import os

adams_links_loc = os.path.abspath(os.getcwd() + '/../data/crawling/adams_links')
f = open(adams_links_loc)
ftxt = f.read()
farr = ftxt.split('\n')

#go through each link name, and find & store the ones that are letters written by Adams.
adamsletts = []
for ind in range(len(farr)):
    linkarr = farr[ind].split(" ")
    firstwds = ' '.join(linkarr[:3])
    if (linkarr[0] == "To") or (firstwds == "John Adams to"):
        adamsletts += [ind]
print len(adamsletts)
#print adamsletts

fout = open('adams_writtenby', 'w')
for ind in adamsletts:
    fout.write('adams_'+str(ind)+'\n')
fout.close()
        
```

```python
#Extract_Senti_Data
#Extract Adams and Washington Sentiment Info from file.
# If you want to change file values, modify "track_sentiment_overtime" ipnb, cells 
#    "WRITE_SENTI_SCORES_WASH" & "WRITE_SENTI_SCORES_ADAMS" [currently, these files are using the
#      "log-wc" metric.]

#washington
wfilenos = []
wwordcounts = []
wnsentis = []
wpsentis = []
fwp = open('wash_letters_sentis_pos')
farr = fwp.read().split('\n')[:-1]
for line in farr:
    larr = line.split(' ')
    wfilenos += [int(larr[0])]
    wwordcounts += [int(larr[1])]
    wpsentis += [float(larr[2])]
fwp.close()
    
fwn = open('wash_letters_sentis_neg')
farr = fwn.read().split('\n')[:-1]
for line in farr:
    larr = line.split(' ')
    wnsentis += [abs(float(larr[2]))]
fwn.close()


#adams
afilenos = []
awordcounts = []
ansentis = []
apsentis = []
fap = open('adams_letters_sentis_pos')
farr = fap.read().split('\n')[:-1]
for line in farr:
    larr = line.split(' ')
    afilenos += [int(larr[0])]
    awordcounts += [int(larr[1])]
    apsentis += [float(larr[2])]
fap.close()
    
fan = open('adams_letters_sentis_neg')
farr = fan.read().split('\n')[:-1]
for line in farr:
    larr = line.split(' ')
    ansentis += [abs(float(larr[2]))]
fan.close()


#DEBUG
"""
print afilenos[:100]
print ""
print awordcounts[:100]
print ""
print apsentis[:100]
print ""
print ansentis[:100]
"""
```

```python
# In this cell, fill in missing date gaps that were in the data. Then print the resulting list of dates into
#   a new file ("adams_fdates_out")
#  [should never need to run this cell again, except maybe to debug.]

fout = open('adams_fdates_out', 'w')
fdates = open('adams_fdates')
farr = fdates.read().split('\n')

lastmonth = 9
lastyear = 1755
nilctr = 0
for line in farr:
    larr = line.split('/')
    if len(larr) != 3:
        continue
    m = int(larr[1])
    y = int(larr[2])
    
    #if there is no month or year, store the date & interpolate it later (and don't do later stuff).
    if (m == -1) or (y == -1):
        nilctr += 1
        continue
        
    #if there is a month or year, interpolate the previous dates and store the current one.
    if (nilctr != 0):
        #find a simple average of this date, and the last recorded date. Do a modulo 12 carry-subtraction
        thisyear = int( (lastyear + y) / 2 )
        if (lastmonth > m):
            thismonth = int( (12 + m - lastmonth)/2 )
            if (thismonth > 12):
                thismonth -= 12
                thisyear += 1
        else:
            thismonth = int( (lastmonth + m)/2 )                
            
        while (nilctr != 0):
            fout.write(str(thismonth) + '/' + str(thisyear) + '\n')
            nilctr -= 1
        #nilctr should be set to zero, so that this doesn't get triggered in the next it.
    
    #make sure the previous month & year indices are updated
    lastmonth = m
    lastyear = y
    
    #write the next date into the file
    fout.write(str(m) + '/' + str(y) + '\n')

#close the read and write files.
fdates.close()
fout.close()            
    
```

```python
# Cell: Asentis_byyear
# Here, compute the mean sentiment for every year for John Adams.
# Also, make lists of senti values for all years in Adams' life (that I have data for).

import numpy

#also, note that very few letters exist before 1775. Therefore, we are only looking at the years
#  between 1775 and 1785.
daterange = range(1775, 1786)


#store years in which all letters were written.
fdates = open('adams_fdates_out')
farr = fdates.read().split('\n')[:-1]
years = [ int(l.split('/')[1]) for l in farr ]
fdates.close()

#find all the letters written by Adams.
f_wby = open('adams_writtenby')
writtenby = f_wby.read().split('\n')[:-1] #contains indices of docs written by adams.
wby_inds = [ int(w.split('_')[1]) for w in writtenby ]

#find sentiment average of all letters written in a given year.
lettinds = {}
ansentis_yr = []
#this contains a collection of all the sentiment values for letters written in all years. Will plot later.
ansentis_byyr = []
for y in daterange:
    #have to store indices in wby, because ansentis only contains wby files.
    lettinds[y] = [ i for i in range(len(wby_inds)) if (years[wby_inds[i]] == y) ]
    ansentis_byyr += [[ ansentis[i] for i in lettinds[y] ]]
    #aggregate all sentiments for the given year.
    y_senti = 0
    for i in lettinds[y]:
        y_senti += ansentis[i]
    ansentis_yr += [ float(y_senti) / len(lettinds[y]) ]
    
#print this for now; plot it later
print ansentis_yr
```
