```python
# This cell: find the letter dates (and store them) in a more efficient way.
#a much more reliable way to collect dates: all of them are displayed in the file "adams_links"!!!
#  unfortunately, these are all off by 1 -____-.
import os

adams_links_loc = os.path.abspath(os.getcwd() + '/../data/crawling/adams_links')
f = open(adams_links_loc)
ftxt = f.read()
farr = ftxt.split('\n')

print "Total number of links: ",
print len(farr)
print ""

#now, go through the file and store the dates in a file.
lett_dates = []
for line in farr:
    linearr = line.split()
    #print linearr[-4:-1]
    lett_dates += [linearr[-4:-1]]

months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October']
months += ['November', 'December']
final_dates = []
for d in lett_dates:
    #try to get the year.
    try:
        #gets rid of brackets in the year.
        ystr = d[2]
        if (ystr[-1] == ']'):
            ystr = ystr[:-1]
        if (ystr[0] == '['):
            ystr = ystr[1:]
            
        year = int(ystr)
        if (year > 1785) or (year < 1755):
            year = -1
    except:
        year = -1
        #debug: just to see what kinds of things cause this.
        #print d
    
    #try to get the month
    if (len(d) > 1):
        month = [m for m in months if (m == d[1])]
        if len(month) == 1:
            mno = months.index(month[0]) + 1
        else:
            mno = -1
    else:
        mno = -1
    
    #try to get the day
    try:
        #gets rid of brackets in the day.
        dstr = d[0]
        if (dstr[-1] == ']'):
            dstr = dstr[:-1]
        if (dstr[0] == '['):
            dstr = dstr[1:]
        
        day = int(dstr)
        if (day > 31) or (day < 0):
            day = -1
    except:
        day = -1
    
    final_dates += [str(day)+'/'+str(mno)+'/'+str(year)]
    
#after we are done, print the final result.
print ""
#print final_dates

#store the final dates in a new file
f = open('adams_fdates', 'w')
for d in final_dates:
    f.write(d+'\n')
f.close()
    
#THE FILE "ADAMS-FDATES" SHOULD NOW HAVE A DATE FOR EVERY FILE (OR AT LEAST MOST FILES)
```

```python
#now, for all the adams files, compute some overall sentiment measure and plot the sentimentality
#  over time.

#do this by opening a file "adams-senti-wds" which contains a sentiment measure for all words,
#  and then for each file, add the sentiment measure for all words in the file. Can mess with this
#  by changing "adams-senti-wds" (in Sentiwordnet_Frequency_analysis ipynb), or can move somewhere else?

#eventually, we may want to incorporate both my measures of "most sentimental words" into this somehow.

import os
from nltk.probability import FreqDist
#these constants are useful for searching through every adams text file.
adams_data_path = os.path.abspath(os.getcwd() + '/../data/crawling/adams_text')

#get the right file names  [before] #   adamsfls = os.listdir(adams_data_path)
f_writtenby = open('adams_writtenby')
adamsfls = f_writtenby.read().split('\n')[:-1]
#print adamsfls



#first, open the file with adams words
fa = open('adams_senti_wds')
fa_txt = fa.read()
fa_posneg = fa_txt.split('=====================')

fa_arr_pos = fa_posneg[0].split('\n')
fa_arr_neg = fa_posneg[1].split('\n')

#read the rest of the words, assign sentiment to each:
adams_senti_wds = {}
for line in fa_arr_pos:
    linearr = line.split()
    if len(linearr) < 3:
        continue
    adams_senti_wds[linearr[1]] = float(linearr[2])

for line in fa_arr_neg:
    linearr = line.split()
    if len(linearr) < 3:
        continue
    adams_senti_wds[linearr[1]] = -1*float(linearr[2])

    
#find the max sentiment, and scale the rest by that one.
max_sent = max(adams_senti_wds.values())
for w in adams_senti_wds:
    adams_senti_wds[w] *= 1/max_sent

afilesentis_pos = []
afilesentis_neg = []
#now, go through each file and come up with a senti score for each
# for now, the senti_score will be the absolute value of the sentiment times the freq of the word.
# later, we can and perhaps should experiment with it more.

#the current metric is multiplying senti scores by word freq within letters.
afilenos = []
awordcounts = []
for fname in adamsfls:
    #record the file name for graphing
    afilenos += [int(fname.split('_')[1])]
    
    f = open(adams_data_path + '/' + fname)
    ftxt = f.read()
    fd = FreqDist(ftxt.split())
    sentisum_pos = 0
    sentisum_neg = 0
    for w in fd:
        if w in adams_senti_wds.keys():
            if adams_senti_wds[w] > 0:
                sentisum_pos += fd[w]*adams_senti_wds[w]
            else:
                sentisum_neg += fd[w]*adams_senti_wds[w]
    #divide the sentiment sum by the total wordcount of the file
    wc = sum(fd.values())
    sentisum_pos *= 1000/wc
    sentisum_neg *= 1000/wc
    afilesentis_pos += [sentisum_pos]
    afilesentis_neg += [sentisum_neg]
    awordcounts += [wc]

```

```python
# CELL: WRITE_SENTI_SCORES_ADAMS.
#requires running the previous cell.
#compute an alternate sentiment measure (# senti words / log-wc), and store it in a file for other analysis.
# This cell only needs to be run once (unless changing letter sentiment metric).
import math
#write positive scores.
fout = open('adams_letters_sentis_pos', 'w')
for i in range(len(afilesentis_pos)):
    score = afilesentis_pos[i] * awordcounts[i] / ( math.sqrt(awordcounts[i]) * 10 )
    #write fileno, then wc, then score
    fout.write(str(afilenos[i]) + ' ' + str(awordcounts[i]) + ' ' + str(score) + '\n')
fout.close()

#write negative scores
fout = open('adams_letters_sentis_neg', 'w')
for i in range(len(afilesentis_neg)):
    score = afilesentis_neg[i] * awordcounts[i] / ( math.sqrt(awordcounts[i]) * 10 )
    #write fileno, then wc, then score
    fout.write(str(afilenos[i]) + ' ' + str(awordcounts[i]) + ' ' + str(score) + '\n')
fout.close()
```

```python
#now, for all the Washington files, compute some overall sentiment measure and plot the sentimentality
#  over time.

#do this by opening a file "wash-senti-wds" which contains a sentiment measure for all words,
#  and then for each file, add the sentiment measure for all words in the file. Can mess with this
#  by changing "wash-senti-wds" (in Sentiwordnet_Frequency_analysis ipynb), or can move somewhere else?

#eventually, we may want to incorporate both my measures of "most sentimental words" into this somehow.

import os
from nltk.probability import FreqDist
#these constants are useful for searching through every adams text file.
wash_data_path = os.path.abspath(os.getcwd() + '/../data/crawling/washington_text')
#get the right file names  [before] #   adamsfls = os.listdir(adams_data_path)
f_writtenby = open('washington_writtenby')
washfls = f_writtenby.read().split('\n')[:-1]




#first, open the file with washington's words
fa = open('wash_senti_wds')
fa_txt = fa.read()
fa_posneg = fa_txt.split('=====================')

fa_arr_pos = fa_posneg[0].split('\n')
fa_arr_neg = fa_posneg[1].split('\n')

#read the rest of the words, assign sentiment to each:
wash_senti_wds = {}
for line in fa_arr_pos:
    linearr = line.split()
    #skips the first line, which is just a title.
    if len(linearr) < 3:
        continue
    wash_senti_wds[linearr[1]] = float(linearr[2])

for line in fa_arr_neg:
    linearr = line.split()
    if len(linearr) < 3:
        continue
    wash_senti_wds[linearr[1]] = -1*float(linearr[2])

    
#find the max sentiment, and scale the rest by that one.
max_sent = max(wash_senti_wds.values())
for w in wash_senti_wds:
    wash_senti_wds[w] *= 1/max_sent

wfilesentis_pos = []
wfilesentis_neg = []
#now, go through each file and come up with a senti score for each
# for now, the senti_score will be the absolute value of the sentiment times the freq of the word.
# later, we can and perhaps should experiment with it more.

#the current metric is multiplying senti scores by word freq within letters.
wfilenos = []
wwordcounts = []
for fname in washfls:
    #record the file name for graphing
    wfilenos += [int(fname.split('_')[1])]
    
    f = open(wash_data_path + '/' + fname)
    ftxt = f.read()
    fd = FreqDist(ftxt.split())
    sentisum_pos = 0
    sentisum_neg = 0
    for w in fd:
        if w in wash_senti_wds.keys():
            if wash_senti_wds[w] > 0:
                sentisum_pos += fd[w]*wash_senti_wds[w]
            else:
                sentisum_neg += fd[w]*wash_senti_wds[w]
    #divide the sentiment sum by the total wordcount of the file
    wc = sum(fd.values())
    sentisum_pos *= 1000/wc
    sentisum_neg *= 1000/wc
    wfilesentis_pos += [sentisum_pos]
    wfilesentis_neg += [sentisum_neg]
    wwordcounts += [wc]
```
